```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definicao da derivada da Lei de Resfriamento de Newton: dy/dt = k * (T_amb - y)
def model(y, k, T_amb):
    return k * (T_amb - y)
```

```python
# Parametros do problema
k = 0.1
T_amb = 20.0
y0 = 100.0
t0 = 0.0
t_final = 50.0
N = 50
```

```python
# Discretizacao do tempo: criamos N intervalos, resultando em N+1 pontos
t = np.linspace(t0, t_final, N + 1)
h = (t_final - t0) / N  # Passo de integracao (delta t)
```

```python
# Inicializacao do array de solucao
y = np.zeros(N + 1)
y[0] = y0
```

```python
# Implementacao do Metodo de Euler
# A aproximacao e dada por: y_{n+1} = y_n + h * f(t_n, y_n)
for i in range(N):
    y[i+1] = y[i] + h * model(y[i], k, T_amb)
```

```python
#plotagem dos resultados
plt.figure(figsize=(10,6))
plt.plot(t, y,'b--',label='Aproximacao Euler')
plt.scatter(t,y,color='red',label='Pontos discretos',zorder=5)
plt.axhline(y=T_amb, color='green', linestyle='-', label='Temperatura Ambiente')

plt.title('Resolução do PVI - Lei de Resfriamento de Newton usando Método de Euler')
plt.xlabel('Tempo (t)')
plt.ylabel('Temperatura (y)')
plt.legend()
plt.grid(True)
plt.show()
```
