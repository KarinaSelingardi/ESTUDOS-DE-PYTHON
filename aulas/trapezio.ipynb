```python
import math 

# definir função f(x)
def f(x):
    return x**2 + math.cos(x)

def regra_trapezio(a, b, N):
    h = (b - a) / N
    soma = f(a) + f(b)
    for i in range(1, N):
        soma += 2 * f(a + i * h)
    integral_aproximada = (h / 2) * soma
    return integral_aproximada

# parâmetros do problema
a = 2.0  # limite inferior
b = 3.0  # limite superior
N = 1000    # número de subintervalos

integral_aproximada = regra_trapezio(a, b, N)
print("Aproximação da integral:", integral_aproximada)


```
