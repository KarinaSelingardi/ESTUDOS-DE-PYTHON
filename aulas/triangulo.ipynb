```python
import math 
def f(x):

    return math.cos(x)  + math.exp(x**2) 

a=2.0
b=3.0
N=10 #INTERVALOS - retângulos 

h= (b-a)/N

soma_alturas = 0.0 

#loop  para calcular a soma das alturas dos retângulos
for i in range(N):
    #calcular a posição do ponto médio do retângulo
    x_med = a + (i + 0.5) * h

    #calcular a altura do retângulo
    altura_retangulo = f(x_med)

    #adicionar a altura à soma
    soma_alturas = soma_alturas + altura_retangulo

#calcular a área total
area_total = soma_alturas * h

print(f"intervalo:[{a}, {b}]")
print(f"número de subintervalos: {N}")
print(f"largura de cada subintervalo: {h}")
print(f"aproximação da integral (regra do ponto médio): {area_total:.5f}")
```

```python
import math 

def f(x):
    return x**2 + math.cos(x)   # aqui usei a função do seu exemplo anterior

def integral_ponto_medio(a, b, N):
    h = (b - a) / N
    soma_alturas = 0.0 

    for i in range(N):
        # ponto médio de cada subintervalo
        x_med = a + (i + 0.5) * h
        soma_alturas += f(x_med)

    area_total = soma_alturas * h
    return area_total

# intervalo
a, b = 2.0, 3.0

# calcular com 100 e 1000 subintervalos
aprox_100 = integral_ponto_medio(a, b, 100)
aprox_1000 = integral_ponto_medio(a, b, 1000)

print(f"Aproximação com 100 subintervalos: {aprox_100:.6f}")
print(f"Aproximação com 1000 subintervalos: {aprox_1000:.6f}")
print(f"Diferença entre elas: {abs(aprox_1000 - aprox_100):.6f}")
```
